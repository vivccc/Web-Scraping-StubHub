```python
import time
from selenium import webdriver
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
```

```python
 def init_driver():
    driver = webdriver.Chrome()
    driver.wait = WebDriverWait(driver, 5)
    return driver
 
def lookup(driver, query):
    driver.get("http://www.google.com")
    try:
        box = driver.wait.until(EC.presence_of_element_located(
            (By.NAME, "q")))
        button = driver.wait.until(EC.element_to_be_clickable(
            (By.NAME, "btnK")))
        box.send_keys(query)
        button.click()
    except TimeoutException:
        print("Box or Button not found in google.com")
 
 
if __name__ == "__main__":
    driver = init_driver()
    lookup(driver, "Selenium")
    time.sleep(5)
    driver.quit()
```

```python
# initialize driver
driver = webdriver.Chrome()
driver.wait = WebDriverWait(driver, 5)
```

```python
driver.get("https://www.stubhub.com/nashville-predators-tickets-nashville-predators-nashville-bridgestone-arena-9-19-2017/event/103043568/?sort=price+asc")
```

```python
# Important processing.
#
# Scrolling to 10 * i th elements for every 1 second -- to display them all
# stop by an error (i.e. all tickets are shown) 
for ticketBoxIndex in range(0, 1000, 5):
    print("ticketBoxIndex %d" % ticketBoxIndex)
    driver.execute_script("$('.rowcell')[" + str(ticketBoxIndex) + "].scrollIntoView(true);")
    time.sleep(1)
```

```python
# all tickets are now shown in #div by the above scrolling, let's get all statistics

ticket_section_elems   = driver.find_elements_by_css_selector('.t-sec')
ticket_row_elems     = driver.find_elements_by_css_selector('.t-row')
ticket_price_elems = driver.find_elements_by_css_selector('.price')
#ticket_num__elems    = driver.find_elements_by_css_selector('.ticketsText')

# Example 
for elem in ticket_price_elems[1:10]:
    elem.text
```

```python
# This transformation is heavy -- takes several minutes
price_list      = [  price.text for   price in ticket_price_elems]
row_list        = [    row.text for     row in ticket_row_elems]
section_list    = [section.text for section in ticket_section_elems]
#ticket_num_list = [    num.text for     num in ticket_num__elems]
```

```python
import pandas as pd

mydf = pd.DataFrame(
    {'price':    price_list,
     'row':        row_list,
     'section':    section_list
     #'ticket_num': ticket_num_list
    })
```
